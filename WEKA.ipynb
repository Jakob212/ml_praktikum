```python
import time
import javabridge
import weka.core.jvm as jvm
from weka.classifiers import Classifier, Evaluation
from weka.core.converters import Loader
from weka.core.classes import Random

jvm.start()
loader = Loader(classname="weka.core.converters.CSVLoader")
data = loader.load_file("your_file.csv")  # Ersetze durch deinen Dateipfad
data.class_is_last()

classifier = Classifier(classname="weka.classifiers.trees.RandomForest")

evaluation = Evaluation(data)
start_time = time.time()
evaluation.crossvalidate_model(classifier, data, 10, Random(1))  #10-Fold-Cross-Validation
end_time = time.time()

print(f"Time taken for cross-validation: {end_time - start_time:.2f} seconds")

jvm.stop()
```
