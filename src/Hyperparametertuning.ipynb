KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Daten laden
data = pd.read_csv('dataset\clf_num\jannis.csv')
X = data.drop('class', axis=1)
y = data['class']

# Hyperparameter-Gitter
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

accuracies = []  # Speichert die Genauigkeiten jeder Iteration

# 20 Durchl채ufe
for i in range(15):
    # Daten splitten
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    
    # Hyperparameter-Tuning
    grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)
    
    # Beste Sch채tzung und Vorhersage
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    
    # Genauigkeit berechnen
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Ergebnisse zusammenfassen
print(f"Durchschnittliche Genauigkeit nach 20 Iterationen: {np.mean(accuracies):.4f}")
print(f"Beste Hyperparameter: {grid.best_params_}")
```

Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Daten laden
data = pd.read_csv('dataset\clf_num\jannis.csv')
X = data.drop('class', axis=1)
y = data['class']

# Hyperparameter-Gitter
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'criterion': ['gini', 'entropy'],
    'max_features': ['auto', 'sqrt', 'log2']
}

accuracies = []

# 20 Durchl채ufe
for i in range(15):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    
    grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)
    
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

print(f"Durchschnittliche Genauigkeit nach 20 Iterationen: {np.mean(accuracies):.4f}")
print(f"Beste Hyperparameter: {grid.best_params_}")
```

Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

# Daten laden
data = pd.read_csv('dataset\clf_num\jannis.csv')
X = data.drop('class', axis=1)
y = data['class']

# Hyperparameter-Gitter
param_grid = {
    'max_depth': [None, 10, 20, 30],
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random']
}

accuracies = []

# 20 Durchl채ufe
for i in range(15):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)
    
    grid = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)
    
    best_model = grid.best_estimator_
    y_pred = best_model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

print(f"Durchschnittliche Genauigkeit nach 20 Iterationen: {np.mean(accuracies):.4f}")
print(f"Beste Hyperparameter: {grid.best_params_}")
```
